```python
from pathlib import Path
import sys


def find_repo_root(start: Path | None = None) -> Path:
    current = (start or Path.cwd()).resolve()
    for candidate in [current, *current.parents]:
        if (candidate / 'lib').exists() and (candidate / 'milestones').exists():
            return candidate
    raise RuntimeError(f'Could not locate repo root from {current}')


repo_root = find_repo_root()
program_dir = repo_root / 'milestones' / 'activation_generic_erf_mf_scaling_convergence'
sys.path.insert(0, str(program_dir))

from semi_empirical_effective_ridge_correction import main

results = main([])
results[:3] if results else results
```
